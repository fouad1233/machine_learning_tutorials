```python
import matplotlib.pyplot as plt
import numpy as np
```

The most important equation used for regression function is as following





### Regression function

```python
def regression(x : np.ndarray , r : np.ndarray, degree : int):
    X = np.array([x**i for i in range(degree )]).T
    #print(X)
    theta = np.linalg.inv(X.T @ (X))@ X.T @ r
    return theta
```

```python
#open the samples.txt file
samples = np.loadtxt("samples.txt")

x = samples[:, 0]
y = samples[:, 1]
plt.title("Veri setindeki x ve y deerleri", fontsize=8)

#set all numbers font at the axis to 8
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
#set the x and y label
plt.xlabel("x", fontsize=8)
plt.ylabel("y", fontsize=8)
plt.scatter(x, y)
plt.show()
```

```python
theta = regression(x, y, 3)

x_new = np.linspace(-30, 20, 100)
y_new = sum([theta[i] * x_new**i for i in range(len(theta))])
# make a title that show the function by using the w and by showing the value of w 3 decimal points
plt.title("f(x) = {:.3f}x^2 + {:.3f}x + {:.3f}".format(theta[2], theta[1], theta[0]), fontsize=8)
#set all numbers font at the axis to 8
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
#set the x and y label
plt.xlabel("x", fontsize=8)
plt.ylabel("y", fontsize=8)
plt.scatter(x, y, )
plt.plot(x_new, y_new, color='red')
#add what color what color means 
plt.legend(["$y_i$", "g(x)"], fontsize=8)
plt.show()

print(' = ', theta)
```
