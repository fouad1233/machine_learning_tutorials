```python
from ml import LinearRegression, regression
import matplotlib.pyplot as plt
import numpy as np

```

```python
#open the samples.txt file
samples = np.loadtxt("samples.txt")

x = samples[:, 0]
y = samples[:, 1]
plt.title("samples points from samples.npy")
plt.scatter(x, y)
plt.show()
```

```python
w = regression(x, y, 3)
print("w: ", w)
x_new = np.linspace(-30, 20, 100)
y_new = sum([w[i] * x_new**i for i in range(len(w))])
# make a title that show the function by using the w and by showing the value of w 3 decimal points
plt.title("f(x) = {:.3f}x^2 + {:.3f}x + {:.3f}".format(w[2], w[1], w[0]))
plt.scatter(x, y)
plt.plot(x_new, y_new, color='red')
#add what color what means
plt.show()
```
