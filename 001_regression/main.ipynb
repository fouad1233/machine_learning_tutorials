```python
from ml import LinearRegression, regression
import matplotlib.pyplot as plt
import numpy as np

```

```python

x = np.array([1, 2, 3, 4, 5])
y = np.array([2.1, 2.9, 4.2, 4.95, 6.05])
```

```python
# plot the data
plt.scatter(x, y)
plt.show()
```

```python
w0, w1 = LinearRegression(x, y)
print("w0: ", w0)
print("w1: ", w1)
```

```python
x_new = np.linspace(0, 6, 100)
y_new = w0 + w1 * x_new

plt.title("Linear Regression")
plt.scatter(x, y)
plt.plot(x_new, y_new, color='red')
plt.show()
```

```python
#try the linear regression of numpy
w1, w0 = np.polyfit(x, y, 1)
print("w0: ", w0)
print("w1: ", w1)
```

```python
#open the samples.txt file
samples = np.loadtxt("samples.txt")

x = samples[:, 0]
y = samples[:, 1]
plt.title("samples points from samples.npy")
plt.scatter(x, y)
plt.show()
```
